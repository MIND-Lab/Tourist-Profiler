```python
import torch
import random
import numpy as np

from time import time
from Setting_DQN import Setting
from Dataset_DQN import Dataset
from Agent_DQN import Agent
from Environment_DQN import Environment
from Utils_DQN import train, test, print_epoch_results, plot_results, save_results

torch.cuda.is_available()
```

# Seed for reproducibility

```python
# Set the random seed for reproducibility.
random_seed = 42
#tf.random.set_random_seed(random_seed)
torch.manual_seed(random_seed)
random.seed(random_seed)
np.random.seed(random_seed)
```

# Profile rating, not separated users and windowed data

## General settings

```python
args = Setting()
args.agent_action_rating_profile = True
args.sep_test_users = False
args.windowed_dataset = True
device = args.device
dataset = Dataset(args.dataset_path, args.batch_size, args.fast_testing)
padding_number = dataset.num_items
agent = Agent(args).to(device)
env = Environment(args, dataset, padding_number, agent.rnn).to(device)

num_epochs = args.num_epochs
```

## Train and evaluation

```python
threshold = 0.0005
prev_rmse = 1000
```

```python
best_reward = -1000

train_loss_list, train_rewards_list = [], []
test_loss_list, test_rewards_list = [], []
rmse_total_list, rmse_explicit_list, rmse_implicit_list = [], [], []
train_times_list, test_times_list = [], []
for epoch in range(0, num_epochs):
    # Train
    train_begin = time()
    train_loss, train_rewards = train(env, agent, args)
    train_loss_list.append(train_loss)
    train_rewards_list.append(train_rewards)
    train_time = time() - train_begin
    train_times_list.append(train_time)
    
    # Test
    test_begin = time()
    test_loss, test_rewards, rmse_total, rmse_explicit, rmse_implicit = test(env, agent, args)
    test_loss_list.append(test_loss)
    test_rewards_list.append(test_rewards)
    rmse_total_list.append(rmse_total)
    rmse_explicit_list.append(rmse_explicit)
    rmse_implicit_list.append(rmse_implicit)
    test_time = time() - test_begin
    test_times_list.append(test_time)
    
    agent.update_epsilon(epoch, num_epochs)
    print_epoch_results(epoch, rmse_total, rmse_explicit, rmse_implicit, 
                        train_rewards, test_rewards,
                        train_time, test_time)

    # if prev_rmse - rmse_total.mean().item() < threshold:
    #     print('Early stopping...')
    #     break
    
    # prev_rmse = rmse_total.mean().item()
    
torch.save(agent.state_dict(), "./models/agent.pth")
torch.save(env.state_dict(), "./models/env.pth")
torch.save(env.rnn.state_dict(), "./models/rnn.pth")

plot_results(train_loss_list, train_rewards_list, 
             test_loss_list, test_rewards_list,
             rmse_total_list, rmse_explicit_list, rmse_implicit_list,
             num_epochs)
save_results(train_loss_list, train_rewards_list, 
             test_loss_list, test_rewards_list,
             rmse_total_list, rmse_explicit_list, rmse_implicit_list,
             args)
print("| Total time: {:5.3f} s. | Train time: {:5.3f} s. | Test time: {:5.3f} s. | ".
      format(np.sum(np.concatenate((train_times_list, test_times_list), axis=0)), 
             np.sum(train_times_list), 
             np.sum(test_times_list)))
```
